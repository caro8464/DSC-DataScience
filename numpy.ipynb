```python
#15
def maxx(x, y):
    """Get the maximum of two items"""
    if x >= y:
        return x
    else:
        return y

# maxx(1, 5) => 5

def pair_max(a,b):
  res = []
  for x in range(a.size):
    if a[x] >= b[x]:
      res.append(a[x])
    else:
      res.append(b[x])
  return np.array(res)

a = np.array([5, 7, 9, 8, 6, 4, 5])
b = np.array([6, 3, 4, 8, 9, 7, 1])

print(pair_max(a, b))

pair_max = np.vectorize(maxx, otypes=[float])

print(pair_max(a,b))
```
